## <span style="color:blue">Otros algoritmos de árboles de clasificación</span>

* ID3 (Iterative Dichotomiser 3) fue desarrollado en 1986 por Ross Quinlan. El algoritmo crea un árbol multidireccional, encontrando para cada nodo (es decir, de manera codiciosa) la característica categórica que generará la mayor ganancia de información para los objetivos categóricos. Los árboles crecen hasta su tamaño máximo y luego se suele aplicar un paso de poda para mejorar la capacidad del árbol para generalizar a datos no vistos.

* C4.5 es el sucesor de ID3 y eliminó la restricción de que las características deben ser categóricas mediante la definición dinámica de un atributo discreto (basado en variables numéricas) que divide el valor del atributo continuo en un conjunto discreto de intervalos. C4.5 convierte los árboles entrenados (es decir, la salida del algoritmo ID3) en conjuntos de reglas si-entonces. Luego se evalúa la precisión de cada regla para determinar el orden en el que deben aplicarse. La poda se realiza eliminando la condición previa de una regla si la precisión de la regla mejora sin ella.

* C5.0 es el lanzamiento de la última versión de Quinlan bajo una licencia propietaria. Utiliza menos memoria y crea conjuntos de reglas más pequeños que C4.5, a la vez que es más preciso.





# <span style="color:blue">Criterios para  subdividir un árbol de clasificación</span>

El criterio para subdividir un árbol en cada paso del algoritmo depende de si se trata de un problema de clasificación (variable objetivo discreta y finita) o de si se trata de un problema de regresión (variable continua o de conteo).

#### <span style="color:#4CC9F0">Indice Gini en el caso dicotómico</span>

La impureza `Gini` es una medida que se utiliza para generar árboles de clasificación. Proporciona más información acerca de la distribución de los datos por nodo que la precisión de clasificación utilizada para informes de precisión del árbol. El índice Gini se calcula con la siguiente expresión:

$$
\text{gini} = \sum_k p_k(1-p_k)
$$




La impureza de un nodo del árbol de clasificación se calcula utilizando el recuento de cada categoría de destino entre todos los registros correspondientes para un nodo dado. El total de impureza Gini se calcula como una suma de cuadrados del recuento de proporciones entre todas las categorías de destino por nodo al que se resta uno y los resultados se multiplican por el número de registros.

Por ejemplo, cuando se divide un nodo de árbol, el algoritmo busca un campo con la mejora más elevada del total de impureza calculada como el total de impureza entre todos los nodos hijo potenciales restados del total de impureza del nodo padre.


Es decir, se suma, para todas las clases, el producto de su proporción por 1 menos su proporción. En el ejemplo si tenemos 5 muestras, 3 de clase 0 y 2 de clase 1, el índice de Gini sería:

$$
\text{gini} = p_0 + p_1 = 3/5 * (1 - 3/5) + 2/5 * (1- 2/5) = 0.48
$$

La distribución de $p * (1 - p)$  es mostrada en la siguiente figura (que ya hemos visto en otras lecciones):


```r
curve(x*(1-x),las=1); 
abline(v=seq(0,1,0.1),h=seq(0,1,0.01),col="gray80") # grilla
```

Es decir, la función toma valores próximos a cero cuando la proporción $p$ de una clase es muy pequeña (lo que implica que el conjunto de valores está muy ordenado, pues si hay pocos elementos de la clase considerada significa que hay muchos de las demás clases) o cuando es muy grande (lo que implica, nuevamente, que hay pocos elementos de las demás clases y que el conjunto está muy ordenado).

El valor máximo de la función anterior se alcanza cuando la proporción es 0.5, lo que implica que, en un escenario con dos clases, el valor máximo del índice de Gini se alcanza en 0.5.

A medida que aumenta el número de clases el valor máximo del índice de Gini aumenta también.

#### <span style="color:#4CC9F0">Indice Gini en el caso de $K$ clases</span>

Si la variable objetivo (target) $Y$ tiene $K$ clases etiquetadas $\{0,1, \ldots, K-1\}$, para el nodo $m$ sean $Q_m$ el subconjunto de datos en el nodo $m$ y $n_m$ el número de elementos en dicho subconjunto. Definimos $p_{mk}$ mediante

$$
p_{mk} =\frac{1}{n_m} \sum_{y\in Q_m} I(y=k). 
$$

Es decir, $p_{mk}$ es la proporción de observaciones de la clase $k$ presentes en el nodo $m$. El `índice de Gini` para el nodo $m$ se define como

$$
H(Q_m) = \sum_{k}p_{mk}(1-p_{mk}).
$$

#### <span style="color:#4CC9F0">Entropía de Shanon</span>

Un segundo criterio de impureza es la entropía de Shanon definida por

$$
H(Q_m) = -\sum_{k}p_{mk}\log p_{mk}.
$$

#### <span style="color:#4CC9F0">Nota aclaratoria</span>

El criterio de entropía calcula la entropía de Shannon de las clases posibles. Toma las frecuencias de clase de los puntos de datos de entrenamiento que alcanzaron una hoja dada como su probabilidad. Usar la entropía de Shannon como criterio de división del nodo del árbol es equivalente a minimizar el logaritmo de la pérdida, también conocida como entropía cruzada o desviación multinomial, entre las etiquetas verdaderas $y_i$ y las predicciones probabilísticas $T_k(x_i)$ del modelo de árbol $T$ para la clase $k$.
Para ver esto, primero recordemos que el log de la pérdida de un modelo de árbol $T$ calculado en un conjunto de datos se define de la siguiente manera:

$$
LL(D, T) = -\frac{1}{n} \sum_{(x_i,y_i)\in D} \sum_{k}I(y_i=k)\log T_k(x_i),
$$

en donde $D$ es un conjunto de entrenamiento con $n$ pares de datos $(x_i, y_i)$.

En un árbol de clasificación, las probabilidades de clase pronosticadas dentro de los nodos hoja son constantes, es decir: para todos los $(x_i,y_i)\in Q_m$, se tiene para cada clase $k$ que $T_k(x_i)=p_{mk}$

Esta propiedad hace posible reescribir $LL(D, T)$ como la suma de las entropías de Shannon calculadas para cada hoja ponderada por el número de puntos de datos de entrenamiento que llegaron a cada hoja:

$$
LL(D, T) =  \sum_{m\in T} \frac{n_m}{n} H(Q_m).
$$

### <span style="color:#4CC9F0">Ventajas de los árboles de decisión</span>

Algunas ventajas de los árboles de decisión son:

1. Fácil de entender y de interpretar. Los árboles se pueden visualizar.
1. Requiere poca preparación de datos. Otras técnicas a menudo requieren la normalización de datos, es necesario crear variables ficticias y eliminar valores en blanco. Sin embargo, tenga en cuenta que el modulo de scikit-learn que usaremos, no admite valores faltantes.
1. El costo de usar el árbol (es decir, la predicción de datos) es logarítmico en la cantidad de puntos de datos usados para entrenar el árbol.
1. Capaz de manejar datos numéricos y categóricos. Sin embargo, la implementación de scikit-learn que usaremos no admite variables categóricas por ahora. Otras técnicas suelen estar especializadas en analizar conjuntos de datos que tienen un solo tipo de variable. 
1. Capaz de manejar problemas de múltiples salidas.
1. Utiliza un modelo de caja blanca. Si una situación dada es observable en un modelo, la explicación de la condición se explica fácilmente mediante lógica booleana. Por el contrario, en un modelo de caja negra (por ejemplo, en una red neuronal artificial), los resultados pueden ser más difíciles de interpretar.
1. Posibilidad de validar un modelo mediante pruebas estadísticas. Eso permite dar cuenta de la fiabilidad del modelo.
1. Tiene un buen desempeño incluso si sus supuestos son algo violados por el verdadero modelo a partir del cual se generaron los datos.

### <span style="color:#4CC9F0">Desventajas de los árboles de decisión</span>

Algunas desventajas de los árboles de decisión son:

1. Los máquinas de aprendizaje de árboles de decisión pueden crear árboles demasiado complejos que no generalizan bien los datos. Esto se llama sobreajuste (overfitting). Para evitar este problema, son necesarios mecanismos como la poda, establecer el número mínimo de muestras requeridas en un nodo de la hoja o establecer la profundidad máxima del árbol.
1. Los árboles de decisión pueden ser inestables porque pequeñas variaciones en los datos pueden generar un árbol completamente diferente. Este problema se mitiga mediante el uso de árboles de decisión dentro de un conjunto.
1. Las predicciones de los árboles de decisión no son uniformes ni continuas, sino aproximaciones constantes por partes. Por lo tanto, no son buenos para hacer extrapolación. 
1. Se sabe que el problema de aprender un árbol de decisión óptimo es NP-completo bajo varios aspectos de optimización e incluso para conceptos simples. En consecuencia, los algoritmos prácticos de aprendizaje del árbol de decisiones se basan en algoritmos heurísticos, como el algoritmo voraz, en el que se toman decisiones localmente óptimas en cada nodo. Dichos algoritmos no pueden garantizar la devolución del árbol de decisión globalmente óptimo. Esto se puede mitigar entrenando varios árboles en un ensamblado, donde las características y las muestras se muestrean aleatoriamente con reemplazo.
1. Hay conceptos que son difíciles de aprender porque los árboles de decisión no los expresan fácilmente, como la función XOR, problemas de paridad o multiplexor.
1. Las máquinas de aprendizaje  de árboles de decisión crean árboles sesgados si dominan algunas clases. Por lo tanto, se recomienda equilibrar el conjunto de datos antes de entrenar el árbol de decisión.

## <span style="color:blue">Arboles de decisión de regresión</span>

Un árbol de regresión es básicamente un árbol de decisión que se usa para la tarea de regresión que se puede usar para predecir salidas de valor continuo en lugar de salidas discretas.

### <span style="color:#4CC9F0">¿Cuál es la predicción de un árbol de regresión?</span>

El valor previsto para el resultado se encuentra al final de las ramas. Para los árboles de regresión, y de forma similar al modelo lineal, el valor predicho corresponderá a la media (o la mediana en caso de variables de conteo) de la región creada por la partición. Para un árbol de clasificación, la predicción corresponderá a una categoría del resultado. 

### <span style="color:#4CC9F0">Criterios para árboles de regresión </span>

Si el objetivo (target) es un valor continuo, entonces para el nodo $m$, los criterios usuales a minimizar para la determinación de ubicaciones para divisiones futuras son el error cuadrático medio (`MSE` o error  `L2`), la `desviación de Poisson` y el `error absoluto medio` (`MAE`  o error `L1`). 

MSE y la desviación de Poisson establecen el valor predicho de los nodos terminales en el `valor medio` aprendido  del nodo mientras que el MAE establece el valor predicho de los nodos terminales en la `mediana`.

#### <span style="color:#4CC9F0">Error cuadrático medio</span>

Esta la función de pérdida preferida cuando la variable objetivo es continua.

$$
\begin{align}
\bar{y} &= \frac{1}{n_m} \sum_{y\in Q_m} y \\
H(Q_m) &= \frac{1}{n_m} \sum_{y\in Q_m} (y-\bar{y}).
\end{align}
$$

#### <span style="color:#4CC9F0">Media desviación de Poisson</span>

Esta es la función de pérdida usual cuando se tiene como variable objetivo a una variable de conteo.

$$
H(Q_m) = \frac{1}{n_m} \sum_{y\in Q_m} \left(y \log \frac{y}{\bar{y}} -y +\bar{y}_m \right)
$$

#### <span style="color:#4CC9F0">Error absoluto medio</span>

Esta función de pérdida es menor que el ECM y está en las mismas unidades que la variable objetivo.


$$
\begin{align}
\text{median}_m &= \text{median}_{y\in Q_m}(y)\\
H(Q_m) &= \frac{1}{n_m} \sum_{y\in Q_m} |y-\text{median}_m|.
\end{align}
$$

## <span style="color:blue">Formulación matemática para la construcción de un árbol de decisión</span>

Matemáticamente el algoritmo se expresa de la siguiente forma. Dado un conjunto de entrenamiento $D$ con parejas de datos $(x_i,y_i)$ un árbol de decisión particiona recursivamente el espacio de características de tal manera que vectores $x_i$ con la misma etiqueta o una similar se clasifican juntos.

Consideremos el nodo $m$ representado por $Q_m$  con $n_m$ muestras. Para cada cada candidato a ser dividido sea $\theta =(j, t_m)$ la pareja que consta de una categoría $j$ y $t_m$ un umbral de separación. Los datos se dividen en dos subconjuntos que denotamos respectivamente por $Q_m^{\text{izq}}$ y $Q_m^{\text{der}}$ y definidos  por


$$
\begin{align}
Q_m^{\text{izq}}(\theta) &= \{(x,y)| x_j<t_m\},\\
Q_m^{\text{der}}(\theta) &= \{(x,y)| x_j\ge t_m\}.
\end{align}
$$
 
La calidad  de un nodo candidato $m$ a ser dividido se calcula usando alguno de los índices de impureza $H()$: Gini, entropía. La escogencia depende de la tarea a ser resuelta: clasificación o regresión.

$$
G(Q_{m}, \theta) = \frac{n_m^{\text{left}}}{n_m}H(Q_m^{left}(\theta))  + \frac{n_m^{\text{right}}}{n_m}H(Q_m^{right}(\theta))
$$

Se seleccionan los parámetros que minimizan la impureza 

$$
\theta^{*} = \text{argmin}_{\theta} G(Q_m, \theta)
$$

recursivamente para los subconjuntos $Q_m^{left}(\theta^*)$ y $Q_m^{right}(\theta^*)$ hasta alcanzar la máxima profundidad, $n_m < \min_{\text{muestras}}$ o $n_m=1$.

## <span style="color:blue">Poda de complejidad de costo mínimo </span>

La poda de complejidad de costo mínimo es un algoritmo utilizado para podar un árbol para evitar el sobreajuste excesivo. El algoritmo es descrito en el Capítulo 3 de [BRE](http://library.lol/main/26908B6EDA02CA4FAF25ADBF57A12B26). Este algoritmo está parametrizado por $\alpha\ge 0$ como el parámetro de complejidad. El parámetro de complejidad se usa para definir la medida de costo-complejidad, $R_{\alpha}(T)$ de un árbol dado: $T$:

$$
R_{\alpha}(T) = R(T) + \alpha|\tilde{T}|,
$$

en dónde $|\tilde{T}|$ es el número de nodos terminales en $T$ y $R(T)$  se define tradicionalmente como la tasa total de clasificación errónea de los nodos terminales. Como se muestra arriba, la impureza de un nodo depende del criterio. La poda de complejidad de costo mínimo encuentra el subárbol de que minimiza $R_{\alpha}(T)$.

La medida de la complejidad del costo de un solo nodo es  $R_{\alpha}(T)= R(t) + \alpha$. La rama,$T_t$, se define como un árbol donde el nodo $t$ es su raíz. En general, la impureza de un nodo es mayor que la suma de las impurezas de sus nodos terminales, $R(T_t)<R(t)$. Sin embargo, la medida de la complejidad del costo de un nodo, $t$, y su rama, $T_t$, puede ser igual dependiendo de $\alpha$. Definimos el $\alpha$ efectivo de un nodo como el valor donde son iguales, $R_{\alpha}(T_t) = R_{\alpha}(t)$ o $\alpha_{\text{eff}} = \tfrac{R(t)-R(T_t)}{|T|-1}$. Un nodo no terminal con el valor más pequeño de $\alpha_{\text{eff}}$ el eslabón más débil y será podado. Este proceso se detiene cuando el $\alpha_{\text{eff}}$ minimsl del árbol podado  es mayor que un  hiperparámetro dado. En scikit-learn ese parámetro es *cc_p_alpha*.
