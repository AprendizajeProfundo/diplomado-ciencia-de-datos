##   <span style="color:blue">R base para gráficos</span>

Una de las partes más poderosas y atractivas de R es su capacidad de crear gráficos de diferentes tipos.

Podemos comenzar, por supuesto, graficando un punto con la función `plot()`. Le tenemos que dar las coordenadas en `X` y `Y`

```r
plot(1,3)
```

Para graficar varios puntos, usamos vectores.

```r
plot(c(1,8), c(10,2))

plot(c(1,2,3,4,5), c(1,2,4,5,6))
```

Podemos también hacer una gráfica a partir de una secuencia

```r
plot(1:200)
```

Y podemos también cambiar la figura que representa los puntos, tanto con `type` como con `pch`

```r
#plot(1:5, type='l')

plot(1:5, pch = 20)
```

Podemos cambiar el tamaño con el argumento `cex`

```r
plot(1:5, pch = 15, cex=5)
```

Por último, cambiamos el color con `col`

```r
plot(1:5, type='l', col="red")
```

si queremos graficar varias lineas en una sola figura, agregamos lineas subsecuentes con la función `lines()`

```r
linea1 <- c(1,2,3,4,5,10)
linea2 <- c(2,5,7,8,9,10)

plot(linea1, type = "l", col = "blue")
lines(linea2, type="l", col = "red")
```

### **Gráficos de dispersión**

La función `plot()` es la que se utiliza para hacer gráficos de dispersión. Hagamos unos datos con correlación para demostrar

```r
set.seed(123)                        
x <- rnorm(30)                       
y <- x + rnorm(30) 

plot(x, y, main="Correlación de muestras", xlab="N", ylab="N + N")
```

Si tenemos grupos o categorías dentro de los datos y queremos representarlos en el gráfico de dispersión, usamos `points()` para agregar otra muestra en la misma figura


```r
x1 <- c(5,7,8,7,2,2,9,4,11,12,9,6)
y1 <- c(99,86,87,88,111,103,87,94,78,77,85,86)

x2 <- c(2,2,8,1,15,8,12,9,7,3,11,4,7,14,12)
y2 <- c(100,105,84,105,90,99,90,95,94,100,79,112,91,80,85)

plot(x1, y1, main="Observaciones en grupos separados", col="red", cex=2)
points(x2, y2, col="blue", cex=2)
```

### **Gráficos de torta**

Los gráficos de torta son perfectos para mostrar tamaños con respecto a la población completa. 
Usamos la función `pie()` y le damos un vector con valores. La torta se separará con respecto al número de valores.

```r
# Creamos un vector con cada uno de los "pedazos"
x <- c(10,20,30,40)

pie(x)
```

Por default cada pedazo se va representando desde el eje x y se mueve contra las manecillas del reloj. Podemos controlar dónde comienza el gráfico con el argumento `init.angle`

```r
pie(x, init.angle = -90)
```

Podemos ponerle etiquetas a los pedazos

```r
labels <- c("Manzanas", "Bananos", "Cerezas", "Dátiles")

pie(x, init.angle = 90, label = labels, main = "Frutas en inventario")
```

Podemos controlar los colores específicos de cada pedazo con el argumento `col`.

```r
colores <- c("green", "red", "blue", "yellow")
pie(x, init.angle = 90, label = labels, main = "Frutas en inventario", col = colores)
```

por último, podemos generar una leyenda a partir de los colores y los títulos de los pedazos con la función `legend()`

```r
labels <- c("Manzanas", "Bananos", "Cerezas", "Dátiles")

colores <- c("green", "red", "blue", "yellow")

pie(x, main = "Gráfico de torta", col = colores)

legend("bottomright", labels, fill = colores)
```

### **Gráficos de barra**

Para obtener gráficos de barra usamos la función `barplot()`. Hay que darle un vector con los valores que van a tomar cada barra y, opcionalmente, el nombre/categoría que se le dará a cada uno de estos.

```r
# eje x: nombres de cada barra
x <- c("A", "B", "C", "D")

# eje y: tamaños de las barras
y <- c(2, 4, 6, 8)

barplot(y, names.arg = x)
```

Podemos cambiar el estilo de las barras. Ya bien sea el color o la textura

```r
barplot(y, names.arg = x, col = "red")

barplot(y, names.arg = x, density = 2)
```

Podemos cambiar los tamaños de cada barra de manera independiente, usando la función `width` y dándole un vector.

```r
barplot(y, names.arg = x, width = c(1,2,3,4))
```

Si le damos datos en forma matricial, se reconocerá inmediatamente como múltiples grupos.

```r
Data <- matrix(1:12, nrow=3, byrow=TRUE)
barplot(Data, names.arg = x, width = c(1,2,3,4))
```

Y para poner los diferentes grupos de lado en lugar de superpuestos, simplemente es cambiar el argumento `beside = TRUE`

```r
barplot(Data, names.arg = x, width = c(1,2,3,4), beside = TRUE,
       col = c("#1b98e0", "#353436", "#0e98b1"), horiz=TRUE)
```

##   <span style="color:blue">GGplot2</span>

Los gráficos básicos de R, aunque poderosos, tienen ciertas desventajas, especialmente en la parte estética. ggplot2 Aparece justamente con la intención de crear fráficas más elegantes y estéticamente agradables.

primero descargamor el paquete y lo llamamos

```r
#install.packages('ggplot2')
library(ggplot2)
```

ggplot está desarrollado para el trabajo en dataframes, así que trabajaremos con un dataset guardado directamente de este paquete

```r
data("midwest", package = "ggplot2")
head(midwest)
```

ahora llamamos la función `ggplot()`, dandole el dataframe como variable y específicando qué columnas queremos utilizar.

```r
ggplot(midwest, aes(x=area, y=poptotal))
```

Pero ¿Qué pasó aquí? Lo que sucede es que ggplot no hace ningún tipo de suposiciones sobre el tipo de gráfico que se quiere. Hasta ahora solo hemos dicho que queremos hacer un gráfico con el dataframe especificado. 

PAra hacer un diagrama de dispersión, "sumamos" un gráfico de puntos a la función inicial.

```r
ggplot(midwest, aes(x=area, y=poptotal)) + geom_point()
```

Ahora ya tenemos un gráfico de dispersión. Pero tenemos varias cosas que agregar para que sea leible.

Por ejemplo, notamos que los límites en los ejes podrían mejorar. Podemos definir  

```r
g <- ggplot(midwest, aes(x=area, y=poptotal)) + geom_point()

g + xlim(c(0, 0.1)) + ylim(c(0, 1000000)) 
```

Este método, como vemos, "borra" algunos datos. Para obtener los mismos resultados podemos hacerle zoom al gráfico.

```r
g = g + coord_cartesian(xlim=c(0,0.1), ylim=c(0, 1000000))
plot(g)
```

Ahora agreguemos un título y cambiemos los nombres de los ejes

```r
g <- g + ggtitle("Area Vs Población", 
            subtitle="Del dataset 'midwest'") + xlab("Area") + ylab("Población")
plot(g)
```

Vamos entendiendo cómo se agregan nuevos elementos a un gráfico de ggplot.

Ahora, esto no se reduce a métodos u opciones estéticas. Podemos definir otros tipos de visualización de esta manera

Agreguemos una linea de ajuste de los datos.

```r
g = g + geom_smooth(method="lm")
plot(g)
```

Ahora podemos ver otros tipos de gráficos que ggplot es capaz de hacer. 

ggplot puede hacer todos los gráficos que vimos en el paquete base, más muchos otros.

### **Violin plot**

Una variación del boxplot que representa los datos de una forma diferente. Un poco más "avant garde"

```r
data(mpg, package="ggplot2")

head(mpg, 5)
```

```r
g <- ggplot(mpg, aes(class, cty))

g + geom_violin() + 
  labs(title="Violin plot", 
       subtitle="Kilometraje por ciudad vs Clase de vehiculo",
       caption="Fuente: mpg",
       x="Clase of Vehicle",
       y="Kilometraje de la ciudad")
```

### **Histogramas**

El paquete base puede hacer histogramas, pero ggplot nos da más control sobre cómo representarlas.

Particularmente nos ayuda a hacer mejores histogramas de variables categóricas

```r
g <- ggplot(mpg, aes(manufacturer))
g + geom_bar(aes(fill=class), width = 0.5) + 
  theme(axis.text.x = element_text(angle=65, vjust=0.6)) + 
  labs(title="Histograma de variables categóricas", 
       subtitle="Manufacturero a través de las clases de autos") 
```

### **Gráficos de densidad**

Tenemos un caso parecido al anterior con los gráficos de densidad

```r
g <- ggplot(mpg, aes(cty))
g + geom_density(aes(fill=factor(cyl)), alpha=0.8) + 
    labs(title="Gráfico de densidad", 
         subtitle="kilometraje agrupado por número de cilíndros",
         caption="Fuente: mpg",
         x="Kilometraje",
         fill="# de cilindros")
```
