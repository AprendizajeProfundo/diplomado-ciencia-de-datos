## <span style="color:#4361EE">Introducción</span>

El pronóstico de series temporales implica la creación de modelos a partir de datos históricos con marca de tiempo para realizar predicciones científicas e impulsar la toma de decisiones estratégicas en el futuro. El pronóstico de series de tiempo tiene muchos usos en varios dominios, que incluyen:

* Predecir la demanda de los consumidores para cada producto
* Pronóstico de la propagación de una pandemia, diagnóstico, medicación y planificación en el cuidado de la salud
* Detección de anomalías, ciberseguridad y mantenimiento predictivo
* Predecir si la infraestructura actual puede manejar el tráfico en el futuro cercano y lejano y muchos más.

El pronóstico de series temporales es un poco diferente del caso de uso de aprendizaje automático tradicional, ya que implica un ordenamiento temporal de los datos que deben tenerse en cuenta durante la ingeniería y el modelado de características.

## <span style="color:#4361EE">Ruido blanco</span>

El término "ruido blanco" es un concepto importante en el análisis de series de tiempo y se refiere a una serie de tiempo en la que los valores son independientes e idénticamente distribuidos (i.i.d) con una media constante y varianza constante. En otras palabras, el ruido blanco es un proceso estocástico en el que los valores en cada período de tiempo son completamente aleatorios y no siguen ningún patrón ni tendencia.

Las principales características del ruido blanco son las siguientes:

* **Media constante**: La media de la serie de tiempo es constante a lo largo del tiempo, lo que significa que, en promedio, los valores no cambian.

* **Varianza constant**e: La varianza de la serie de tiempo también es constante, lo que implica que la dispersión de los valores es constante en cada período.

* **Independencia temporal**: Los valores en un momento dado son independientes de los valores en otros momentos. No hay correlación o dependencia entre los valores sucesivos.

* **Distribución normal**: Los valores siguen una distribución normal o gaussiana. Sin embargo, en la práctica, no siempre se cumple estrictamente este requisito.

Sin patrón ni tendencia: No hay patrones discernibles ni tendencias en la serie de tiempo. Los valores son impredecibles y no siguen ninguna estructura específica.

El ruido blanco es útil en el análisis de series de tiempo como un punto de referencia para comparar con otros modelos. Cuando se ajusta un modelo a una serie de tiempo, es importante verificar si los residuos del modelo (diferencia entre los valores observados y los valores predichos) se comportan como ruido blanco. Si los residuos se asemejan al ruido blanco, esto sugiere que el modelo está capturando adecuadamente la estructura de los datos.

En la práctica, se utilizan pruebas estadísticas para verificar si los residuos de un modelo son ruido blanco. Una prueba común es la Prueba de Ljung-Box, que evalúa si los residuos son independientes a través del tiempo.

Con ayuda de chat-GPT.

Un proceso de ruido blanco se suele escribir matemáticamente como una sucesión $\{ a_t\}_{t=1,2,\cdots}$, en donde las $a_t$ son variables aleatoria indpendientes con $a_t \sim N(0, \sigma_a^2)$




```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 

sigma_a = 1.0
mu_a = 0

# ruido blanco
rb = pd.DataFrame(norm.rvs(size=1000, loc=mu_a, scale=sigma_a))
rb.columns=["a_t"]
rb.head()

# plot
rb.a_t.plot(figsize = (20,5))
plt.title('Ruido blanco N(0,1)')
plt.xlabel('t')
plt.show()
```

## <span style="color:#4361EE">Polinomio de retraso</span>

Supongamos que $W_t$ representa una serie de tiempo. Definimos el retraso de $W_t$ mediante

$$
B W_t = W_{t-1}
$$

Observe que 

$$
B^2 W_t = B (BW_t) = W_{t-2}
$$

Si $\phi_1, \phi_2, \cdots, \phi_p$ son números reales, el polinomio de retraso $\phi(B)$ se define mediante

$$
(1- \phi_1 B - \phi_2 B^2 - \cdots \phi_p B^p) W_t
$$

Para una costante $\mu$ se tiene que $B\mu = \mu$.

## <span style="color:#4361EE">Operador de diferenciación</span>

Para una serie de tiempo $w_t$ el operador de diferenciación $\nabla$ se define mediante

$$
\nabla W_t = w_t - w_{t-1}
$$

Para una constante $\mu$ se tiene que $\nabla \mu = 0$.

Adicionalmente se define

$$
\nabla^{d+1} = \nabla \nabla^{d}
$$

## <span style="color:#4361EE">Serie autoregresiva de orden 1: AR(1)</span>

Supongamos que $0< \phi <1$ es una constante y $\mu$ un número real. Supongamos que $W_t$ representa un proceso aleatorio descrito mediante la ecuación

$$
(1-\phi B) (W_t-\mu) = a_t
$$
en donde $a_t$ es ruido blanco.


El proceso puede reescribirse como

$$
W_t = + \mu(1-\phi) + \phi W_{t-1} + a_t
$$

Si $W_0 = 0, \mu=0$ y $\phi= 0.6$ se obtiene la serie

```python
r = np.array(rb.a_t)
w = np.zeros_like(r)
phi = 0.6
```

```python
# construye la serie
z = [phi*w[t-1] + r[t] for t in range(1, r.shape[0])]
```

```python
w_ar = pd.DataFrame(z)
w_ar.columns = ['w_t']
w_ar.shape
```

```python
# plot
w_ar.w_t.plot(figsize = (20,5))
plt.title('Proceso autoregresivo AR(1)')
plt.xlabel('t')
plt.show()
```

## <span style="color:#4361EE">Serie autoregresiva de orden p: AR(p)</span>

La generalizacion de un proceso AR(1) es bastante simple

Supogamos que $\phi(B)$ es un polinomio de retraso de grado p, es decir $\phi(B)= (1- \phi_1 B - \phi_2 B^2 - \cdots \phi_p B^p)$. Un proceso o serie autoregresiva de orden $p$ es dado por

$$
\phi(B) (W_t-\mu) = a_t
$$
en donde $a_t$ es ruido blanco.

## <span style="color:#4361EE">Serie Promedio móvil de orden 1: MA(1)</span>

Supongamos que $0< \theta <1$ es una constante y $\mu$ un número real. Supongamos que $Z_t$ representa un proceso aleatorio descrito mediante la ecuación

$$
 Z_t-\mu = (1-\theta B)a_t
$$
en donde $a_t$ es ruido blanco.


El proceso puede reescribirse como

$$
Z_t =  \mu + a_t -\theta a_{t-1}
$$

Si $ \mu=0$ y $\theta= 0.8$ se obtiene la serie

```python
r = np.array(rb.a_t)
z_t = np.zeros_like(r)
theta = 0.8
```

```python
# construye la serie
z_t = [ r[t]- theta* r[t-1]  for t in range(1, r.shape[0])]
```

```python
z_ma = pd.DataFrame(z_t)
z_ma.columns = ['z_t']
z_ma.shape
```

```python
# plot
z_ma.z_t.plot(figsize = (20,5))
plt.title('Proceso promedio móvil MA(1)')
plt.xlabel('t')
plt.show()
```

## <span style="color:#4361EE">Serie promedio móvil de orden q: MA(q)</span>

La generalizacion de un proceso AR(1) es bastante simple

Supogamos que $\theta(B)$ es un polinomio de retraso de grado q, es decir $\theta(B)= (1- \theta_1 B - \theta_2 B^2 - \cdots \theta_q B^q)$. Un proceso o serie de promedio móvil  de orden $q$ es dado por

$$
Z_t-\mu = \theta(B) a_t
$$
en donde $a_t$ es ruido blanco.

## <span style="color:#4361EE">Modelo Arma: ARMA(p,q)</span>

A las combinaciones de los modelas AR y MA se les conoce con el nombre de modelos autoregresivos de promedios máviles (ARMA) y definen genéricamente mediante la ecuación 

$$
\phi(B)(Z_t-\mu) = \theta(B) a_t 
$$

## <span style="color:#4361EE">Modleo ARIMA(p,d,q)</span>

Haciendo uso de los polinomios de retraso y del operador diferneica se obtienen los modelos ARIMA 


$$
\phi(B)\nabla^d(Z_t) = \theta(B) a_t 
$$



## <span style="color:#4361EE">Estacionariedad</span>

La estacionariedad es una parte clave del análisis de series de tiempo. En pocas palabras, la estacionariedad significa que la forma en que cambian los datos de las series temporales es constante. Una serie temporal estacionaria no tendrá tendencias ni patrones estacionales. Debe verificar la estacionariedad porque no solo facilita el modelado de series de tiempo, sino que también es una suposición subyacente en muchos métodos de series de tiempo. Específicamente, se asume estacionariedad para una amplia variedad de métodos de pronóstico de series temporales, incluidos el promedio móvil autorregresivo (ARMA), ARIMA y ARIMA estacional (SARIMA).

Usaremos la prueba de Dickey Fuller para verificar la estacionariedad de nuestros datos. La Prueba de Dickey-Fuller busca determinar la existencia o no de raíces unitarias en una serie de tiempo. La hipótesis nula de esta prueba es que existe una raíz unitaria en la serie.


Esta prueba generará valores críticos y un valor $p$, que nos permitirá aceptar o rechazar la hipótesis nula de que no existe estacionariedad. Si rechazamos la hipótesis nula, eso significa que aceptamos la alternativa, que establece que existe estacionariedad. 
Estos valores nos permiten probar el grado en que los valores presentes cambian con los valores pasados. Si no hay estacionariedad en el conjunto de datos, un cambio en los valores actuales no provocará un cambio significativo en los valores pasados.

Para detalles de la prueba puede revisar en [Wikipedia](https://es.wikipedia.org/wiki/Prueba_de_Dickey-Fuller).


Probemos la estacionariedad en los datos de pasajeros de nuestras aerolíneas. Para empezar, calculemos una media móvil (rolling mean) de siete meses:

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adft = adfuller(z_ma,autolag="AIC")

output_df = pd.DataFrame({"Values":[adft[0],adft[1],adft[2],adft[3], adft[4]['1%'], adft[4]['5%'], adft[4]['10%']]  , "Metric":["Test Statistics","p-value","No. of lags used","Number of observations used", 
                                                        "critical value (1%)", "critical value (5%)", "critical value (10%)"]})
print(output_df)
```

### <span style="color:#4CC9F0">Interpretación del resultado de la prueba.</span>

Para cualquier nivel de significancia la hipótesis null "la serie es estacionaria" es rechazada contundentemente. Es decir la serie no es estacionaria.
